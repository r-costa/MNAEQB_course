Material Balances:

$$
\begin{aligned}
&\frac{\mathrm{dVCD}}{\mathrm{dt}}=μ \cdot VCD  \quad  \text{(exponential growth)}\\
&\frac{\mathrm{dSub}}{\mathrm{dt}}=-v_s⋅VCD + \text{feed_rate} \quad (v_s\text{ incorporates both growth-associated and maintenance-associated consumption)}
\end{aligned}
$$

Reaction Kinetics:

$$
\begin{align}
\mu &= \frac{\mathrm{Sub}}{\mathrm{Sub+6.1}} \quad \text{(μ - specific growth rate, given by Monod equation)}\\
v_s &= (\alpha \cdot \mu + \beta \cdot \text{Sub}) \quad \text{(substrate consumption rate; α=0.65, β=0.075 - empirical parameters)}
\end{align}
$$
